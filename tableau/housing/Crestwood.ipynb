```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/NewCrestwood.csv")

df.head()
```

```python
df['Date_Sold'] = df['Date_Sold'].str.replace(',', '', regex=True)

df['Price'] = df['Price'].str.replace(',', '', regex=True).str.replace('$', '', regex=True)

df['Square_Feet'] = df['Square_Feet'].str.replace(',', '', regex=True)
```

```python
df.head()
```

```python
df['Price'] = df['Price'].str.replace('M', '0000', regex=True).str.replace('.', '', regex=True)
```

```python
df.dtypes
```

```python
# df.astype({'col1': 'int32'}).dtypes

df = df.astype({
    'Date_Sold': 'datetime64[ns]',
    'Price': 'int32',
    'Square_Feet': 'int32'
})
```

```python
df.to_csv("datasets/updated_crestwood.csv")
```
